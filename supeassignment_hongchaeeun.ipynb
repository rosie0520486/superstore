```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
file_path = '/content/drive/My Drive/Sample - Superstore.csv'
df = pd.read_csv(file_path, encoding='latin1')
print(df.head())
```

```python
columns_of_interest = ['Region', 'Category', 'Sub-Category', 'Customer Name', 'Sales', 'Quantity', 'Profit']
df = df[columns_of_interest]
grouped = df.groupby(['Region', 'Category', 'Sub-Category', 'Customer Name']).agg(
    Total_Sales=('Sales', 'sum'),
    Total_Quantity=('Quantity', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Avg_Sales=('Sales', 'mean'),
    Avg_Quantity=('Quantity', 'mean'),
    Avg_Profit=('Profit', 'mean')
).reset_index()
print(grouped.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
region_grouped = df.groupby('Region').agg(
    Total_Sales=pd.NamedAgg(column='Sales', aggfunc='sum'),
    Average_Sales=pd.NamedAgg(column='Sales', aggfunc='mean'),
    Total_Profit=pd.NamedAgg(column='Profit', aggfunc='sum'),
    Average_Profit=pd.NamedAgg(column='Profit', aggfunc='mean'),
    Total_Quantity=pd.NamedAgg(column='Quantity', aggfunc='sum'),
    Average_Quantity=pd.NamedAgg(column='Quantity', aggfunc='mean')
).reset_index()
print(region_grouped)
plt.figure(figsize=(14, 8))
sns.set_theme(style="whitegrid")
sns.barplot(x='Region', y='Total_Sales', data=region_grouped, palette='Blues', label='Total Sales', alpha=0.8)
sns.barplot(x='Region', y='Total_Profit', data=region_grouped, palette='Oranges', label='Total Profit', alpha=0.8, hatch="//")
plt.legend(loc='upper left')
plt.title('Region-wise Total Sales and Profit')
plt.show()
```

```python
file_path = '/content/drive/My Drive/Sample - Superstore.csv'
df = pd.read_csv(file_path, encoding='latin1')
columns_of_interest = ['Region', 'Category', 'Sub-Category', 'Customer Name', 'Sales', 'Quantity', 'Profit','Discount','Product Name'] # Changed 'product name' to 'Product Name'
df = df[columns_of_interest]
south_tables = df[(df['Region'] == 'South') & (df['Category'] == 'Furniture') & (df['Sub-Category'] == 'Tables')]
product_discount = south_tables.groupby('Product Name')['Discount'].mean().reset_index()
high_discount_products = product_discount[product_discount['Discount'] > product_discount['Discount'].mean()]
high_discount_customers = south_tables[south_tables['Product Name'].isin(high_discount_products['Product Name'])]['Customer Name'].value_counts().reset_index()
high_discount_customers.columns = ['Customer Name', 'Purchase Count']
print("Products with high discount rates in the South region (Tables):")
print(high_discount_products)
print("\nCustomers who frequently purchased high-discount tables in the South region:")
print(high_discount_customers)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.barplot(data=product_discount, x='Product Name', y='Discount', palette='muted')
plt.title('average discount rate for sales of tables in south(product)')
plt.xlabel('product name')
plt.ylabel('average discountrate')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
west_tables = df[(df['Region'] == 'West') & (df['Category'] == 'Furniture') & (df['Sub-Category'] == 'Tables')]
product_discount_west = west_tables.groupby('Product Name')['Discount'].mean().reset_index()
high_discount_products_west = product_discount_west[product_discount_west['Discount'] > product_discount_west['Discount'].mean()]
high_discount_customers_west = west_tables[west_tables['Product Name'].isin(high_discount_products_west['Product Name'])]['Customer Name'].value_counts().reset_index()
high_discount_customers_west.columns = ['Customer Name', 'Purchase Count']

print("Products with high discount rates in the West region (Tables):")
print(high_discount_products_west)
print("\nCustomers who frequently purchased high-discount tables in the West region:")
print(high_discount_customers_west)

plt.figure(figsize=(12, 6))
sns.barplot(data=product_discount_west, x='Product Name', y='Discount', palette='muted')
plt.title('Average discount rate for sales of tables in West (product)')
plt.xlabel('Product Name')
plt.ylabel('Average Discount Rate')
plt.xticks(rotation=90)
plt.show()
```

```python
south_tables = df[(df['Region'] == 'South') & (df['Category'] == 'Furniture') & (df['Sub-Category'] == 'Tables')]
product_profit_ratio = south_tables.groupby('Product Name')['Profit'].sum() / south_tables.groupby('Product Name')['Sales'].sum()
product_profit_ratio = product_profit_ratio.reset_index()
product_profit_ratio.columns = ['Product Name', 'Profit Ratio']
plt.figure(figsize=(12, 6))
sns.barplot(data=product_profit_ratio, x='Product Name', y='Profit Ratio', palette='muted')
plt.title('Profit Ratio of Tables in South Region by Product')
plt.xlabel('Product Name')
plt.ylabel('Profit Ratio')
plt.xticks(rotation=90)
plt.show()
low_profit_products = product_profit_ratio[product_profit_ratio['Profit Ratio'] < product_profit_ratio['Profit Ratio'].mean()]
print("Products with low profit ratio in the South region (Tables):")
print(low_profit_products)
region_category_sales = df.groupby(['Region', 'Category'])['Sales'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=region_category_sales, x='Region', y='Sales', hue='Category', palette='muted')
plt.title('Sales by Region and Category')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.legend(title='Category')
plt.show()
```

```python
region_category_sales = df.groupby(['Region', 'Category'])['Sales'].sum().reset_index()
region_total_sales = region_category_sales.groupby('Region')['Sales'].sum().reset_index()
region_category_sales = pd.merge(region_category_sales, region_total_sales, on='Region', suffixes=('', '_Total'))
region_category_sales['Sales_Ratio'] = region_category_sales['Sales'] / region_category_sales['Sales_Total']
import matplotlib.pyplot as plt
import numpy as np
regions = region_category_sales['Region'].unique()
categories = region_category_sales['Category'].unique()
fig, ax = plt.subplots(figsize=(12, 6))
width = 0.8 / len(categories)
bottom = np.zeros(len(regions))
for i, category in enumerate(categories):
    values = region_category_sales[region_category_sales['Category'] == category]['Sales_Ratio']
    ax.bar(regions, values, width, bottom=bottom, label=category)
    bottom += values
    ax.set_ylabel('Sales Ratio')
ax.set_title('Sales Ratio by Category and Region')
ax.legend(title='Category')
plt.xticks(regions)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
file_path = '/content/drive/My Drive/Sample - Superstore.csv'
df = pd.read_csv(file_path, encoding='latin1')
columns_of_interest = ['Region', 'Category', 'Sub-Category', 'Customer Name', 'Sales', 'Quantity', 'Profit','Discount','Product Name','Ship Mode']
df = df[columns_of_interest]
south_tables = df[(df['Region'] == 'South') & (df['Category'] == 'Furniture') & (df['Sub-Category'] == 'Tables')]
correlation_data = south_tables[['Sales', 'Discount', 'Quantity']] # Ship Mode는 범주형 변수이므로 제외
correlation_matrix = correlation_data.corr()
print(correlation_matrix)
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Sales, Discount, and Quantity for Tables in South Region')
plt.show()
plt.figure(figsize=(10, 6))
sns.boxplot(x='Ship Mode', y='Sales', data=south_tables)
plt.title('Sales by Ship Mode for Tables in South Region')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Ship Mode', y='Discount', data=south_tables)
plt.title('Discount by Ship Mode for Tables in South Region')
plt.show()
```

```python
import pandas as pd

# 배송 모드 숫자형 변환
ship_mode_mapping = {
    'Standard Class': 1,
    'Second Class': 2,
    'First Class': 3,
    'Same Day': 4
}
df['Ship Mode Num'] = df['Ship Mode'].map(ship_mode_mapping)
correlation_data = df[['Ship Mode Num', 'Sales', 'Discount', 'Quantity']]
correlation_matrix = correlation_data.corr()
print(correlation_matrix)
import seaborn as sns
import matplotlib.pyplot as plt

# 히트맵 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Ship Mode, Sales, Discount, and Quantity')
plt.show()
```

```python
tables_df = df[(df['Category'] == 'Furniture') & (df['Sub-Category'] == 'Tables')]
# 지역별 Ship Mode 사용 빈도 계산 (tables_df 사용)
ship_mode_freq_region = tables_df.groupby(['Region', 'Ship Mode'])['Ship Mode'].count().reset_index(name='Frequency')

# 지역별 전체 빈도 계산 (tables_df 사용)
region_total_freq = ship_mode_freq_region.groupby('Region')['Frequency'].sum().reset_index()

# 지역별 Ship Mode 사용 비율 계산 (tables_df 사용)
ship_mode_freq_region = pd.merge(ship_mode_freq_region, region_total_freq, on='Region', suffixes=('', '_Total'))
ship_mode_freq_region['Proportion'] = ship_mode_freq_region['Frequency'] / ship_mode_freq_region['Frequency_Total']
fast_ship_modes = ['First Class', 'Same Day']
fast_ship_mode_freq_region = ship_mode_freq_region[ship_mode_freq_region['Ship Mode'].isin(fast_ship_modes)]
fast_ship_mode_freq_region = fast_ship_mode_freq_region.groupby('Region')['Proportion'].sum().reset_index()
fast_ship_mode_freq_region.rename(columns={'Proportion': 'Fast Ship Mode Proportion'}, inplace=True)
# 누적 막대 그래프 시각화 (tables_df 기반)
plt.figure(figsize=(12, 6))
sns.barplot(data=ship_mode_freq_region, x='Region', y='Proportion', hue='Ship Mode', palette='muted', dodge=False)
plt.title('Ship Mode Frequency by Region (Tables)')
plt.xlabel('Region')
plt.ylabel('Proportion')
plt.legend(title='Ship Mode')
plt.show()

# 막대 그래프 시각화 (빠른 Ship Mode 사용 비율, tables_df 기반)
plt.figure(figsize=(10, 6))
sns.barplot(data=fast_ship_mode_freq_region, x='Region', y='Fast Ship Mode Proportion', palette='muted')
plt.title('Fast Ship Mode (First Class, Same Day) Proportion by Region (Tables)')
plt.xlabel('Region')
plt.ylabel('Proportion')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate ship mode frequencies for each region
ship_mode_freq_region = df.groupby(['Region', 'Ship Mode'])['Ship Mode'].count().reset_index(name='Frequency')

# Calculate total frequency for each region
region_total_freq = ship_mode_freq_region.groupby('Region')['Frequency'].sum().reset_index()

# Merge total frequency back to ship_mode_freq_region
ship_mode_freq_region = pd.merge(ship_mode_freq_region, region_total_freq, on='Region', suffixes=('', '_Total'))

# Calculate proportion of each ship mode within each region
ship_mode_freq_region['Proportion'] = ship_mode_freq_region['Frequency'] / ship_mode_freq_region['Frequency_Total']

# Visualize using a stacked bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=ship_mode_freq_region, x='Region', y='Proportion', hue='Ship Mode', palette='muted', dodge=False)
plt.title('Ship Mode Frequency by Region')
plt.xlabel('Region')
plt.ylabel('Proportion')
plt.legend(title='Ship Mode')
plt.show()
```
